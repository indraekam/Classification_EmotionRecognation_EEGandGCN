```python
import numpy as np
import scipy.signal as sps
import matplotlib.pyplot as plt

def frame_based_fft(signal, frame_size, hop_size, fs, freq_range=(4, 45), window_type='hann'):
    signal_length = len(signal)
    if frame_size > signal_length:
        raise ValueError("Frame size cannot be larger than the signal length.")

    num_frames = (signal_length - frame_size) // hop_size + 1
    frames = np.zeros((num_frames, frame_size), dtype=np.complex128)

    # Buat jendela (window)
    window = sps.get_window(window_type, frame_size)

    for i in range(num_frames):
        start = i * hop_size
        end = start + frame_size

        frame = signal[start:end]

        # Terapkan jendela pada frame
        frame = frame * window

        # Lakukan FFT pada frame
        frame_fft = np.fft.fft(frame)

        # Simpan hasil FFT frame
        frames[i] = frame_fft

    # Hitung sumbu frekuensi
    frequency_axis = np.fft.fftfreq(frame_size, d=1/fs)
    freq_mask = (frequency_axis >= freq_range[0]) & (frequency_axis <= freq_range[1])
    frequency_axis = frequency_axis[freq_mask]

    return frames[:, freq_mask], frequency_axis

# # Contoh penggunaan pemrosesan berbasis frame dengan FFT dan windowing pada sinyal single kanal
# signal = np.random.randn(1000)  # Sinyal acak dengan panjang 1000
# frame_size = 200  # Ukuran frame
# hop_size = 100  # Langkah (overlap) antara frame
# fs = 200  # Frekuensi sampel (Hz)
# freq_range = (4, 45)  # Rentang frekuensi (Hz)

# frames, frequency_axis = frame_based_fft(signal, frame_size, hop_size, fs, freq_range)

# # Cetak hasil FFT frame pertama
# print("FFT Frame Pertama:", frames[0])
# print("Shape FFT Frame Pertama:", frames[0].shape)

# # Tampilkan frekuensi
# plt.plot(frequency_axis, np.abs(frames[0]))
# plt.xlabel('Frekuensi (Hz)')
# plt.ylabel('Amplitudo')
# plt.title('FFT Frame Pertama')
# plt.grid(True)
# plt.show()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

class GCN(nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super(GCN, self).__init__()
        self.conv1 = nn.Conv2d(in_channels, hidden_channels, kernel_size=1)
        self.conv2 = nn.Conv2d(hidden_channels, out_channels, kernel_size=1)

    def forward(self, x, edge_index):
        x = self.conv1(x)
        x = F.relu(x)
        x = self.conv2(x)
        return x

# Data EEG representation
# Anda perlu mengubah representasi data EEG menjadi bentuk grafik (misalnya adjacency matrix / edge_index).

# Contoh sederhana: kita anggap setiap elektroda pada EEG merepresentasikan simpul dalam grafik, dan ada koneksi (edge) antara elektroda yang berdekatan.

num_electrodes = 12  # Jumlah elektroda pada data EEG
adjacency_matrix = torch.zeros(num_electrodes, num_electrodes)  # Buat matriks ketetanggaan (adjacency matrix)

# Misalnya, kita anggap setiap elektroda terhubung dengan dua elektroda tetangganya pada kiri dan kanan (dalam kasus ini, elektroda 0 dan 9 bersambung).
for i in range(num_electrodes):
    if i == 0:
        adjacency_matrix[i, 1] = 1
    elif i == num_electrodes - 1:
        adjacency_matrix[i, num_electrodes - 2] = 1
    else:
        adjacency_matrix[i, i - 1] = 1
        adjacency_matrix[i, i + 1] = 1

# Ubah adjacency matrix menjadi tensor PyTorch
edge_index = adjacency_matrix.nonzero(as_tuple=False).t()

# Data EEG input
# Misalnya, kita memiliki data EEG dengan 10 kanal (elektroda) dan panjang sinyal adalah 100.

num_channels = 12
signal_length = 1000

# Buat data acak sebagai contoh
eeg_data = torch.tensor(x_train, dtype = torch.float)

print(len(eeg_data))

# Bentuk data masukan menjadi bentuk (batch_size, in_channels, signal_length, 1)
# Dimensi ke-4 adalah 1 karena kita menggunakan operasi konvolusi 2D.
x = eeg_data.unsqueeze(-1)

print(x.shape)

# Inisialisasi model GCN
in_channels = num_channels
hidden_channels = 32
out_channels = 3
gcn_model = GCN(in_channels, hidden_channels, out_channels)

# Latih model GCN
learning_rate = 0.01
num_epochs = 100

optimizer = optim.Adam(gcn_model.parameters(), lr=learning_rate)
criterion = nn.CrossEntropyLoss()

# label = []
# for i in range(len(eegdatalatih)):
#   label.append(eegdatalatih[i][1])


labels =  torch.tensor(y_train)
for epoch in range(num_epochs):
    optimizer.zero_grad()
    output = gcn_model(x, edge_index)
    # Lakukan operasi pooling (misalnya, global average pooling) untuk mendapatkan output akhir dari GCN.
    output = F.adaptive_avg_pool2d(output, (1, 1)).squeeze()

    # Misalnya, kita memiliki 5 kelas untuk klasifikasi
    # labels = torch.randint(3, (1,)).long()
    # print(output)
    loss = criterion(output, labels)
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}')

# Setelah melatih model, Anda dapat menggunakannya untuk prediksi pada data EEG baru.
# Misalnya, Anda dapat melakukan forward pass dengan data baru dan mendapatkan prediksi kelasnya.
```

```python
# Gunakan model yang telah dilatih untuk melakukan prediksi
with torch.no_grad():
    gcn_model.eval()  # Set model dalam mode evaluasi (tidak memperbarui parameter)
    output = gcn_model(x, edge_index)
    output = output.squeeze(-1)
    output = F.adaptive_avg_pool1d(output, 1).view(-1, out_channels)
    predicted_class_idx = torch.argmax(output, dim=1)

# predicted_class = predicted_class_idx.item()
print(f'Predicted class index: {predicted_class_idx}')
```
